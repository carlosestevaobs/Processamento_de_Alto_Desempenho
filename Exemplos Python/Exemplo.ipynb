```python
import time
import multiprocessing
import cv2
import numpy as np
from multiprocessing.pool import ThreadPool as Pool
from matplotlib import pyplot as plt


def transformar(imagem):    
    for i in range (int(len(imagem))):
        if (imagem[i] == 0):
            imagem[i] = 255
    return imagem
    

if __name__ == '__main__':
    processos = int(input("Digite a quantidade de processos que desejar "))
    startT = time.time()
    pool = multiprocessing.Pool(processes=processos)   
    imagem = cv2.imread("imagemTeste.jpg", 0)   
    
    NovaImagem = pool.map(transformar, (imagem))
        
    endT = time.time()   
    
    print("Tempo: "+ str(endT - startT)+" segundos")
    
```

```python
plt.subplot(121)
plt.title("Imagem original")
plt.imshow(imagem, cmap="gray")
        
plt.subplot(122)
plt.title("Resultado")
plt.imshow(NovaImagem, cmap="gray")
plt.show()
```
